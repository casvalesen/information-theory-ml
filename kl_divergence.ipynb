## KL Divergence 

- Example of KL divergence calculation

```python
import numpy as np
from scipy.stats import expon
from scipy.stats import entropy
import matplotlib.pyplot as plt

# Generate random samples from the first exponential distribution
np.random.seed(42)

scale_1 = 1
scale_2 = 1.2

samples1 = expon.rvs(scale=scale_1, size=1000)

# Generate random samples from the second exponential distribution with slightly different parameter
samples2 = expon.rvs(scale=scale_2, size=1000)

# Calculate the true PDF values of the two distributions
x = np.linspace(0, np.max(samples1 + samples2), 1000)
pdf1 = expon.pdf(x, scale=scale_1)
pdf2 = expon.pdf(x, scale=scale_2)


# Compute the KL divergence between the two distributions
kl_divergence = entropy(pdf1, pdf2)

# Plot the two distributions
plt.figure(figsize=(8, 5))
plt.hist(samples1, bins=30, alpha=0.5, label='Distribution 1')
plt.hist(samples2, bins=30, alpha=0.5, label='Distribution 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Exponential Distributions')
plt.legend()

plt.show()

print("KL Divergence:", kl_divergence)
```
