# Information theory Basics 

- This notebook demonstrates some information theory basics 

## Self information 

- Self information of an event happening is defined as $ I(x) = - log P(x) $
- The less probability for an event, the more information is contained in the event happening. 

```python
x = np.linspace(0.01, 1, 100 )
I = - np.log(x)
plt.plot(x, I)
```

```python
x = np.linspace(0.0, 1, 100 )
I = np.log(x)
plt.plot(x, I)
```
